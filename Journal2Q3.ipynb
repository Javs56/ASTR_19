```python
import matplotlib.pyplot as plt
import numpy as np
#setting range of 0-1
a = [0,1]
for i in a:
    #uniform distribution of 1000 numbers
    a = np.random.uniform(0,1,1000)
bins = 100 #100 groups for histogram
plt.hist(a,bins,alpha=0.5,edgecolor="black")
plt.xlabel("100 Groups From 0 to 1")
plt.ylabel("Magnitude")
plt.savefig("Journal2Q3.pdf", bbox_inches="tight",facecolor="white")
```
