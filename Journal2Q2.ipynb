```python
import matplotlib.pyplot as plt
import numpy as np

x = np.arange(0,1,0.01) #set x as an array from 0-1 in increments of 100

def sin(): #returns sin of x
    for i in x:
        return np.sin(x)
    
def cos(): #returns cos of x
    for i in range(0,1,0.01):
        return np.cos(x)
```

```python
y = np.sin(x) #y value = sin
y1 = np.cos(x) #other y value = cos
f, ax = plt.subplots(1,2) #setting multipanel plot
ax[0].plot(x, y) #0 is the first panel
ax[0].set_title("sin(x)")
ax[1].plot(x, y1) #1 is the second panel
ax[1].set_title("cos(x)")
plt.savefig("Journal2Q2.pdf", bbox_inches="tight",facecolor="white")
```
