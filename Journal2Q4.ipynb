```python
import matplotlib.pyplot as plt
import numpy as np

#setting range 0-1
a = [0,1]
for i in a:
    #permutation distribution of 1000 numbers
    a = np.random.permutation(1000)
bins = 100 #100 bins for histogram
plt.hist(a,bins,alpha=0.5,edgecolor="black")
plt.xlabel("100 Groups From 0 to 1")
plt.ylabel("Magnitude")
plt.savefig("Journal2Q4.pdf", bbox_inches="tight",facecolor="white")
```
